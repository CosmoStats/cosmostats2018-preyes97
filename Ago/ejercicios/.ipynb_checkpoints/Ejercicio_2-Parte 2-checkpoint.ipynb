 Distribuciones de probabilidad 
 De las siguientes distribuciones calcular: 
 1-CDF
 2-Valor medio
 3-Promedio
 4-Varianza

Distribución Gaussiana:
\begin{equation}
p(x|\mu \sigma) = \frac{1}{\sigma \sqrt{2 \pi}}exp(-\frac{(x-\mu)^2}{2\sigma^2})
\end{equation}

1-CDF:
\begin{equation}
F(x) = \int_{-\infty}^{x} f(x') dx' = \frac{1}{\sigma \sqrt{\pi}}\int_{-\infty}^{x} exp(-\frac{(x-\mu)^2}{2 \sigma^2}) dx 
 \end{equation}

Realizando un cambio de variable:
\begin{eqnarray}
t = \frac{(x-\mu)}{\sigma}&&\\
dt = \frac{dx}{\sigma}
\end{eqnarray}

Se tiene pues: 

\begin{equation}
F(x) = \frac{\sigma}{\sigma \sqrt{\pi}} \int_{-\infty}^{x} exp(-\frac{t^2}{2}) dt  
\end{equation}

Se concluye: 
\begin{equation}
F(x) = \frac{1}{\sqrt{\pi}}\int_{-\infty}^{x}exp(-\frac{t^2}{2})
\end{equation}

2-Valor medio:

\begin{equation}
E(x) = \int_{-\infty}^{\infty} xf(x) dx 
\end{equation}

\begin{equation}
\int_{-\infty}^{\infty} x \frac{1}{\sigma \sqrt{2 \pi}} exp(-\frac{(x-\mu)^2}{2 \sigma})dx = \int_{- \infty}^{\infty}(x + \mu) \frac{1}{\sigma \sqrt{2 \pi}} exp(-\frac{x^2}{2\sigma^2}) dx
\end{equation}

\begin{equation}
= \int_{-\infty}^{\infty} \frac{x}{\sigma \sqrt{2 \pi}}exp(-\frac{(x)^2}{2 \sigma}) dx + \int_{-\infty}^{\infty} \frac{\mu}{\sigma \sqrt{2 \pi}}exp(-\frac{(x)^2}{2 \sigma}) dx
\end{equation}

Separando la primera integral: 
\begin{equation}
\int_{-\infty}^{0} \frac{x}{\sigma \sqrt{2 \pi}}exp(-\frac{(x)^2}{2 \sigma}) dx + \int_{0}^{\infty} \frac{x}{\sigma \sqrt{2 \pi}}exp(-\frac{(x)^2}{2 \sigma}) dx
\end{equation}

\begin{equation}
-\int_{0}^{-\infty} \frac{x}{\sigma \sqrt{2 \pi}}exp(-\frac{(x)^2}{2 \sigma}) dx + \int_{0}^{\infty} \frac{x}{\sigma \sqrt{2 \pi}}exp(-\frac{(x)^2}{2 \sigma}) dx
\end{equation}

\begin{equation}
\int_{0}^{\infty} \frac{-x}{\sigma \sqrt{2 \pi}}exp(-\frac{(-x)^2}{2 \sigma}) dx + \int_{0}^{\infty} \frac{x}{\sigma \sqrt{2 \pi}}exp(-\frac{(x)^2}{2 \sigma}) dx
\end{equation}

\begin{equation}
-\int_{0}^{\infty} \frac{x}{\sigma \sqrt{2 \pi}}exp(-\frac{(x)^2}{2 \sigma}) dx + \int_{0}^{\infty} \frac{x}{\sigma \sqrt{2 \pi}}exp(-\frac{(x)^2}{2 \sigma}) dx
\end{equation}

$$= 0 $$

Entonces: 
\begin{equation}
E(x) = \int_{-\infty}^{\infty} \frac{\mu}{\sigma \sqrt{2 \pi}} exp(-\frac{x^2}{2 \sigma^2}) dx
\end{equation}

\begin{equation}
= \frac{2 \mu}{\sqrt{\pi}} \int_{-\infty}^{\infty} exp^{-x^2} dx
\end{equation}

Multiplicando por $$\sigma \sqrt{2}$$

\begin{equation}
E(x) = \int_{-\infty}^{\infty} \frac{\mu}{\sqrt{\pi}} exp(-x^2) dx = \frac{2 \mu}{\sqrt{\pi}}\int_{0}^{\infty}exp(-x^2) dx
\end{equation}

Ahora: 
\begin{equation}
\frac{2}{\sqrt{\pi}} \int_{0}^{\infty} exp(-x^{2}) dx = \lim_{x -> \infty} \frac{2}{\sqrt{\pi}} \int_{0}^{t} exp(-x^2) dx 
\end{equation}

\begin{equation}
\lim_{t -> \infty} erf(t) = 1 
\end{equation}

Por lo tanto: 
$$E(x) = \mu$$

3-Varianza
\begin{equation}
Var(x) = \int_{-\infty}^{\infty} \frac{(x-\mu)^2}{\sigma \sqrt{2\pi}}exp(-\frac{(x-\mu)^2}{2 \sigma^2}dx
\end{equation}

\begin{equation}
= \int_{-\infty}^{\infty} \frac{(x-\mu)^2}{\sigma \sqrt{2 \pi}}exp(-\frac{(x-\mu)^2}{2 \sigma^2} dx
\end{equation}

\begin{equation}
= \int_{-\infty}^{\infty}\frac{x^2}{\sigma \sqrt{2\pi}} exp(-\frac{-x^2}{2 \sigma^2})dx
\end{equation}

\begin{equation}
= \sigma\sqrt{2} \int_{-\infty}^{\infty}(\sigma \sqrt{2}x)^2\frac{1}{\sigma \sqrt{2\pi}}exp(-\frac{(\sigma \sqrt{2}x)^2}{2 \sigma^2}) dx
\end{equation}

\begin{equation}
= \sigma^{2}\frac{4}{\sqrt{\pi}} \int_{0}^{\infty}x^2 exp(-x^2) dx 
\end{equation}

Realizando un cambio de variable: 
\begin{eqnarray}
t = x^2 && dt = 2xdx = 2\sqrt{t}dx &&\\
x = t && dx = (2\sqrt{t}^-1 dt &&\\
\end{eqnarray}

Sustituyendo:
\begin{equation}
V(x) = \sigma^{2}\frac{4}{\sqrt{\pi}} \int_{0}^{\infty}(\sqrt{t})^2(2\sqrt{2})^-1 exp(-t)dt
\end{equation}

\begin{equation}
= \sigma^2 \frac{4}{\sqrt{\pi}}\frac{1}{2}\int_{0}^{\infty} t^{\frac{3}{2}-1}exp(-1) dt 
\end{equation}

\begin{equation}
= \sigma^2 \frac{4}{\pi}\frac{1}{2}\Gamma(\frac{3}{2})
\end{equation}

\begin{equation}
V(x) = \sigma^{2}\frac{4}{\sqrt{\pi}}\frac{1}{2}\frac{\sqrt{\pi}}{2} = \sigma^2
\end{equation}




2-Distribución uniforme

\begin{eqnarray}
p(x|\mu, w) = \frac{1}{w} para |x-\mu| \leq \frac{w}{2}&&\\
w = b-a
\end{eqnarray}


1-CDF:

\begin{equation}
F(x) = \int_{-\infty}^{a} f(x') dx 
\end{equation}

Si $$x \leq a$$

\begin{equation}
F(x) = \int_{- \infty}^{0} dx = \int_{-\infty}^{0} \frac{1}{b-a} dx = 0
\end{equation}

Si $$a \leq x \lt 0$$
\begin{equation}
F(x) = \int_{a}^{x} f(x') dx' = \int_{a}^{x} \frac{dx'}{b-a} 
\end{equation}

\begin{equation}
\frac{1}{b-a}\int_{a}^{x} dx' = \frac{x-a}{b-a}
\end{equation}

Si $$ x \gt b$$

\begin{equation}
F(x) = \int_{a}^{b} f(x') dx' = \int_{a}^{b} \frac{dx}{b-a}
\end{equation}

\begin{equation}
= \frac{1}{b-1}\int_{a}^{}b dx' = 1
\end{equation}

Media:
\begin{equation}
E[X] = \int_{-\infty}^{infty} x f(x) dx  
\end{equation}

Nos restringimos a $$[a,b]$$

\begin{equation}
E[x] = \int_{a}^{b} \frac{x}{b-a}dx = \frac{(b^2 + a^2)}{2(b-a)} = \frac{b+a}{2}
\end{equation}

Varianza: 

\begin{equation}
Var(x) = E[x^2] = - (E[x])^2
\end{equation}

\begin{equation}
\int_{-\infty}^{\infty} f(x) x^2 dx - (\frac{b+a}{2})^2
\end{equation}

\begin{equation}
\int_{a}^{b} \frac{x^2}{b-a} dx -  \frac{(b+a)^2}{4}
\end{equation}

\begin{equation}
= \frac{b^3 - a^3}{(b-a)^3} - \frac{(b+a)^2}{4} = \frac{(b-a)^2)}{12}
\end{equation}








Distribución binomial 
\begin{equation}
    \binom{n}{k} = \frac{n!}{k!(n-k)!}
\end{equation}


\begin{equation}
    f(k,n,p) = \binom{n}{k}p^k(1-p)^{(n-k)}
\end{equation}

1-CDF:
\begin{equation}
F(x̣̣_{n}) = F(cx_{n}) =  \sum_{x_{i}<x_{o}} f(x_{i})
\end{equation}


\begin{equation}
F(x;n;p) = \sum_{i=0}^{x} \frac{n!}{(n-1)!k!} p^{i}(1-p)^{(n-1)}
\end{equation}


2-Media:
\begin{equation}
E(x) = \sum_{x = 0}^{n} x ( \frac{n!}{(n-x)!x!}p^{x}(1-p)^{(n-x)}
\end{equation}

\begin{equation}
= \sum_{x = 1}^{n} \left[ \frac{n!}{(n-x)! x!}\right] p^{x}(1-p)^{(n-p)}
\end{equation}

\begin{equation}
= \sum_{x = 1}^{n} \left[ \frac{n!}{(n-1) - (x-1)! (x-1)!}\right] p^{x}(1-p)^{(n-p)}
\end{equation}

Cambio: $$i = x-1$$

\begin{equation}
E(x) = n \sum_{i=0}^{n-1} \left[ \frac{(n-1)!}{((n-1)-i)! (i)!} \right] p^{i-1}(1-p)^{n-1}
\end{equation}

\begin{equation}
= np \sum_{i=0}^{n-1} \left[ \frac{(n-1)!}{((n-1)-i)! (i)!}\right] p^{i}(1-p)^{[(n-1)-i]}
\end{equation}

\begin{equation}
= np (p+(1-p))
\end{equation}

\begin{equation}
E(x) = np
\end{equation}

3-Varianza:

\begin{equation}
Var[x] = E[x^2]-(E[x])^2 = E[x^2] - n^2p^2
\end{equation}

\begin{equation}
E[x^2] = \sum_{x = 0}^{n} x^2 \left( \frac{n!}{(n-x)!n!} \right) p^{n} (1-p)^(n-x)
\end{equation}

\begin{equation}
E[x^2] = [x(x-1)+x] \left[ \frac{n!}{(n-x)!n!} \right] p^{n} (1-p)^(n-x)
\end{equation}

\begin{equation}
= \sum_{x=2}^{n} \left[ \frac{n!}{(n-x)!(x-2)!\right] p^{x}(1-p)^{n-x} + np
\end{equation}

\begin{equation}
= np + n(n-1)p^{2}p^{2}\sum_{x=2}^{n} \left[ \frac{n!}{(n-x)!(x-2)!\right] p^{x-2}(1-p)^{n-k}
\end{equation}

Con $$k = x - 2$$

\begin{equation}
= np + n(n-1)p^{2}[p + (1-p)]^{n-2}
\end{equation}

\begin{equation}
E[x^2] = np + n(n-1)p^2 = np + (n^2 - n)p^2 = np + (np)^2 - np^2
\end{equation}

\begin{equation}
Var[x] = np(1-p)
\end{equation}
