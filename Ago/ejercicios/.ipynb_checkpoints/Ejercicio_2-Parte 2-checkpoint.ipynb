Distribución Gaussiana:
\begin{equation}
p(x|\mu \sigma) = \frac{1}{\sigma \sqrt{2 \pi}}exp(-\frac{(x-\mu)^2}{2\sigma^2})
\end{equation}

1-CDF:
\begin{equation}
F(x) = \int_{-\infty}^{x} f(x') dx' = \frac{1}{\sigma \sqrt{\pi}}\int_{-\infty}^{x} exp(-\frac{(x-\mu)^2}{2 \sigma^2}) dx 
 \end{equation}

Realizando un cambio de variable:
\begin{eqnarray}
t = \frac{(x-\mu)}{\sigma}&&\\
dt = \frac{dx}{\sigma}
\end{eqnarray}

Se tiene pues: 

\begin{equation}
F(x) = \frac{\sigma}{\sigma \sqrt{\pi}} \int_{-\infty}^{x} exp(-\frac{t^2}{2}) dt  
\end{equation}

Se concluye: 
\begin{equation}
F(x) = \frac{1}{\sqrt{\pi}}\int_{-\infty}^{x}exp(-\frac{t^2}{2})
\end{equation}

2-Valor medio:

\begin{equation}
E(x) = \int_{-\infty}^{\infty} xf(x) dx 
\end{equation}

\begin{equation}
\int_{-\infty}^{\infty} x \frac{1}{\sigma \sqrt{2 \pi}} exp(-\frac{(x-\mu)^2}{2 \sigma})dx = \int_{- \infty}^{\infty}(x + \mu) \frac{1}{\sigma \sqrt{2 \pi}} exp(-\frac{x^2}{2\sigma^2}) dx
\end{equation}

\begin{equation}
= \int_{-\infty}^{\infty} \frac{x}{\sigma \sqrt{2 \pi}}exp(-\frac{(x)^2}{2 \sigma}) dx + \int_{-\infty}^{\infty} \frac{\mu}{\sigma \sqrt{2 \pi}}exp(-\frac{(x)^2}{2 \sigma}) dx
\end{equation}

Separando la primera integral: 
\begin{equation}
\int_{-\infty}^{0} \frac{x}{\sigma \sqrt{2 \pi}}exp(-\frac{(x)^2}{2 \sigma}) dx + \int_{0}^{\infty} \frac{x}{\sigma \sqrt{2 \pi}}exp(-\frac{(x)^2}{2 \sigma}) dx
\end{equation}

\begin{equation}
-\int_{0}^{-\infty} \frac{x}{\sigma \sqrt{2 \pi}}exp(-\frac{(x)^2}{2 \sigma}) dx + \int_{0}^{\infty} \frac{x}{\sigma \sqrt{2 \pi}}exp(-\frac{(x)^2}{2 \sigma}) dx
\end{equation}

\begin{equation}
\int_{0}^{\infty} \frac{-x}{\sigma \sqrt{2 \pi}}exp(-\frac{(-x)^2}{2 \sigma}) dx + \int_{0}^{\infty} \frac{x}{\sigma \sqrt{2 \pi}}exp(-\frac{(x)^2}{2 \sigma}) dx
\end{equation}

\begin{equation}
-\int_{0}^{\infty} \frac{x}{\sigma \sqrt{2 \pi}}exp(-\frac{(x)^2}{2 \sigma}) dx + \int_{0}^{\infty} \frac{x}{\sigma \sqrt{2 \pi}}exp(-\frac{(x)^2}{2 \sigma}) dx
\end{equation}

$$= 0 $$

Entonces: 
\begin{equation}
E(x) = \int_{-\infty}^{\infty} \frac{\mu}{\sigma \sqrt{2 \pi}} exp(-\frac{x^2}{2 \sigma^2}) dx
\end{equation}

\begin{equation}
= \frac{2 \mu}{\sqrt{\pi}} \int_{-\infty}^{\infty} exp^{-x^2} dx
\end{equation}

Multiplicando por $$\sigma \sqrt{2}$$

\begin{equation}
E(x) = \int_{-\infty}^{\infty} \frac{\mu}{\sqrt{\pi}} exp(-x^2) dx = \frac{2 \mu}{\sqrt{\pi}}\int_{0}^{\infty}exp(-x^2) dx
\end{equation}

Ahora: 
\begin{equation}
\frac{2}{\sqrt{\pi}} \int_{0}^{\infty} exp(-x^{2}) dx = \lim_{x -> \infty} \frac{2}{\sqrt{\pi}} \int_{0}^{t} exp(-x^2) dx 
\end{equation}

\begin{equation}
\lim_{t -> \infty} erf(t) = 1 
\end{equation}

Por lo tanto: 
$$E(x) = \mu$$

3-Varianza

\begin{equation}

Realizando un cambio de variable:

t = x^2 && dt = 2xdx = 2\sqrt{t}dx &&\\

\end{eqnarray}

Sustituyendo:
\begin{equation}
V(x) = \sigma^{2}\frac{4}{\sqrt{\pi}} \int_{0}^{\infty}(\sqrt{t})^2(2\sqrt{2})^-1 exp(-t)dt

2-Distribución uniforme

\begin{eqnarray}
p(x|\mu, w) = \frac{1}{w} para |x-\mu| \leq \frac{w}{2}&&\\
w = b-a
\end{eqnarray}


1-CDF:

\begin{equation}
F(x) = \int_{-\infty}^{a} f(x') dx 
\end{equation}

Si $$x \leq a$$

\begin{equation}
F(x) = \int_{- \infty}^{0} dx = \int_{-\infty}^{0} \frac{1}{b-a} dx = 0
\end{equation}

Si $$a \leq x \lt 0$$
\begin{equation}
F(x) = \int_{a}^{x} f(x') dx' = \int_{a}^{x} \frac{dx'}{b-a} 
\end{equation}

\begin{equation}
\frac{1}{b-a}\int_{a}^{x} dx' = \frac{x-a}{b-a}
\end{equation}

Si $$ x \gt b$$

\begin{equation}
F(x) = \int_{a}^{b} f(x') dx' = \int_{a}^{b} \frac{dx}{b-a}
\end{equation}

\begin{equation}
= \frac{1}{b-1}\int_{a}^{}b dx' = 1
\end{equation}

Media:
\begin{equation}
E[X] = \int_{-\infty}^{infty} x f(x) dx  
\end{equation}

Nos restringimos a $$[a,b]$$

\begin{equation}
E[x] = \int_{a}^{b} \frac{x}{b-a}dx = \frac{(b^2 + a^2)}{2(b-a)} = \frac{b+a}{2}
\end{equation}

Varianza: 

\begin{equation}
Var(x) = E[x^2] = - (E[x])^2
\end{equation}

\begin{equation}
\int_{-\infty}^{\infty} f(x) x^2 dx - (\frac{b+a}{2})^2
\end{equation}

\begin{equation}
\int_{a}^{b} \frac{x^2}{b-a} dx -  \frac{(b+a)^2}{4}
\end{equation}

\begin{equation}
= \frac{b^3 - a^3}{(b-a)^3} - \frac{(b+a)^2}{4} = \frac{(b-a)^2)}{12}
\end{equation}








Distribución binomial

    \binom{n}{k} = \frac{n!}{k!(n-k)!}
\end{equation}


\begin{equation}
    f(k,n,p) = \binom{n}{k}p^k(1-p)^{(n-k)}
\end{equation}

1-CDF:

F(x̣̣_{n}) = F(cx_{n}) =  \sum_{x_{i}<x_{o}} f(x_{i})

\begin{equation}
F(x;n;p) = \sum_{i=0}^{x} \frac{n!}{(n-1)!k!} p^{i}(1-p)^{(n-1)}
\end{equation}


2-Media:
\begin{equation}
E(x) = \sum_{x = 0}^{n} x ( \frac{n!}{(n-x)!x!}p^{x}(1-p)^{(n-x)}
\end{equation}

\begin{equation}
= \sum_{x = 1}^{n} \left[ \frac{n!}{(n-x)! x!}\right] p^{x}(1-p)^{(n-p)}
\end{equation}

\begin{equation}
= \sum_{x = 1}^{n} \left[ \frac{n!}{(n-1) - (x-1)! (x-1)!}\right] p^{x}(1-p)^{(n-p)}
\end{equation}

Cambio: $$i = x-1$$

\begin{equation}
E(x) = n \sum_{i=0}^{n-1} \left[ \frac{(n-1)!}{((n-1)-i)! (i)!} \right] p^{i-1}(1-p)^{n-1}
\end{equation}

\begin{equation}
= np \sum_{i=0}^{n-1} \left[ \frac{(n-1)!}{((n-1)-i)! (i)!}\right] p^{i}(1-p)^{[(n-1)-i]}

\begin{equation}
= np (p+(1-p))

\begin{equation}

3-Varianza:

\begin{equation}
Var[x] = E[x^2]-(E[x])^2 = E[x^2] - n^2p^2
\end{equation}

E[x^2] = \sum_{x = 0}^{n} x^2 \left( \frac{n!}{(n-x)!n!} \right) p^{n} (1-p)^(n-x)
\end{equation}

\begin{equation}
= [x(x-1)+x] \left[ \frac{n!}{(n-x)!n!} \right] p^{n} (1-p)^(n-x)

\begin{equation}

Distribución de Poisson

\begin{equation}
f(K, \lambda) = e^{-\lambda} \frac{\lambda^k}{k!} 
\end{equation}

1-CDF: 
\begin{equation}
F(k,\lambda) = \sum_{x=0}^{n} e^{-\lambda} \frac{\lambda^{x}}{x!} = e^{-\lambda}\sum_{x=0}^{k} \frac{\lambda^x}{x!}
\end{equation}

\begin{equation}
P(x < \lambda) = F(k,\lambda) = e^{-\lambda}\sum_{x=0}^{k} \frac{\lambda^x}{x!}
\end{equation}

2-Media:
\begin{equation}
E[x] = \sum_{x=0}^{\infty} x e^{-\lambda}\frac{\lambda^x}{x!}
\end{equation}

\begin{equation}
E[x] = \sum_{x=1}^{\infty} xe^{-\lambda}\frac{\lambda^x}{x!} = \sum_{x = 1}^{\infty} x e^{-\lambda} \frac{(\lambda) \lambda^{x-1}}{(x)(x-1)!}
\end{equation}

\begin{equation}
= \lambda e^{-\lambda} \sum_{x = 1}^{\infty}\frac{\lambda^{x-1}}{(x-1)!}
\end{equation}

COn el cambio: $$k = x-1 $$

\begin{equation}
= \lambda e^{-\lambda} \sum_{k=0}^{\infty}\frac{\lambda^k}{(k)!} = \lambda e^{-\lambda}e^{\lambda} = \lambda
\end{equation}

3-Varianza: 
$$Var[x] = E[x^2] - (E[x])^2$$
Donde $$E[x^2] = E[x(x-1) +x] = E[x(x-1)] + E[x]$$

\begin{equation}
E[x(x-1)] =  \sum_{x=0}^{\infty}(x(x-1))e^{-\lambda} \frac{\lambda^{x}}{x!} 
\end{equation}

\begin{equation}
= \sum_{x=0}^{\infty}\frac{e^{-\lambda}\lambda^{x}}{(x-2)!}
\end{equation}

\begin{equation}
= \sum_{x=2}^{\infty} e^{-\lambda^2}\frac{\lambda^{x-2}}{(x-2)!} = \lambda^{2}e^{-\lambda}\sum_{x=2}^{\infty}\frac{\lambda^{x-2}}{(x-2)!} 
\end{equation}

Con el cambio: $$k = x - 2$$

\begin{equation}
E[x(x-1)] = \lambda^{2}e^{-\lambda}\sum_{k=0}^{\infty}\frac{\lambda^k}{k!} = \lambda^{2}
\end{equation}

\begin{equation}
Var[x] = E[x^2] - (E[x])^2 = E[x(x+1)] + E[x] - (E[x])^2 = \lambda^2 + \lambda - \lambda^{2}
\end{equation}











Distribución beta:

\begin{equation}
F(x) = \frac{(x-a)^{p-1}(b-x)^{q-1}}{B(p,q)(b-a)^{p+q-1}}
\end{equation}

Donde: 
\begin{eqnarray}
a \leq x \leq b; && p,q > 0   
\end{eqnarray}

1-CDF:
\begin{equation}
F(x) = \frac{1}{B(p,q)}\int_{0}^{x} x'^{p-1} (1-x')^{q-1}dx' 
\end{equation}

2-Media: 
\begin{equation}
E[x] = \int_{0}^{1} xf(x;\alpha,\beta) dx
\end{equation}

\begin{equation}
= \int_{0}^{1} x \frac{x^{p-1}(1-x)^{\beta - 1}dx}{B(p,q)}
\end{equation}

\begin{equation}
= \frac{B(p+1,q)}{B(p,q)}
\end{equation}

\begin{equation}
= \frac{\Gamma(p + 1)\Gamma(q) \Gamma(p+q)}{\Gamma(p + q + 1) \Gamma(\alpha)\Gamma(\beta)} = \frac{p}{p+q}
\end{equation}


3-Varianza:
\begin{equation}
Var(x) = \int_{0}^{1} x^{2}f(x) dx - (E[x])^2
\end{equation}

\begin{eqnarray}
Var(x) = \frac{1}{B(p,q)}\int_{0}^{1} x^{p+1}(1-x)^{q-1}dx - \frac{p^2}{(p+q)^2} \\
= \frac{B(p+2,q)}{B(p,q)} - \frac{p^2}{(p+q)^2}\\
= \frac{\Gamma(p+2)\Gamma(q)}{\Gamma(p+q+2)} \frac{\Gamma(p+q)}{\Gamma(p)\Gamma(q)} - \frac{p^2}{(p+q)^2} \\ 
= \frac{pq}{(p+q)^2(p+q+1)}
\end{eqnarray}



Distribución de Bernoulli 

\begin{equation}
f(x) = \frac{1}{2^{\frac{n}{2}}\Gamma(\frac{n}{2})} x^{\frac{n}{2}-1} e^{-\frac{x}{2}} 
\end{equation}

1-CDF:

\begin{equation}
F(X) = \int_{0}^{x} \frac{1}{2^{2 \frac{n}{2}} \Gamma(\frac{n}{2})} x'^{\frac{n}{2}-1} e^{-\frac{x'}{2}} dx 
\end{equation}

\begin{equation}
 = \frac{1}{2\Gamma(\frac{n}{2})}\int_{0}^{x}(\frac{x'}{2})^{(\frac{n}{2})-1}e^{-\frac{x'}{2}} dx'
\end{equation}

Para ahorrar procedimiento se aconseja utilizar la función gamma incompleta: 

\begin{equation}
\gamma(s,x) = \int_{x}^{\infty} t^{s-1}e^{-t}dt
\end{equation}

Entonces, si realizamos un cambio de variable $$v = \frac{x'}{2}$$  
\begin{equation}
= \frac{1}{\Gamma(\frac{n}{2})} \int_{0}^{x} v^{(\frac{n}{2})-1}e^{-v} dv'
\end{equation}

Se obtiene pues: 
\begin{equation}
= \frac{\gamma(\frac{n}{2},\frac{x}{2})}{\Gamma(\frac{n}{2})}
\end{equation}

2-Media:

Para simplificar realizamos una sustitución: $$  a = \frac{1}{ 2^{\frac{n}{2}} \Gamma(\frac{n}{2})}  $$

\begin{equation}
E[x] = \int_{0}^{\infty} xf(x) dx = \int_{0}^{\infty}xax^{\frac{x}{2}-1}  e^{\frac{x}{2}} dx
\end{equation}

\begin{equation}
= a \int_{0}^{\infty}x^{\frac{x}{2}}  e^{\frac{x}{2}} dx 
\end{equation}


\begin{equation}
= a \left[-x^{\frac{n}{2}} 2 exp(-\frac{x}{2}) \right] + \int]_{0}^{\infty} n x^{\frac{x}{2}-1} exp(-\frac{x}{2}) dx
\end{equation}

\begin{equation}
= c\left[ n \int_{0}^{\infty}  x^{\frac{x}{2}-1} exp(-\frac{x}{2}) dx\right] 
\end{equation}

\begin{equation}
= n \int_{0}^{\infty}xf(x)dx 
\end{equation}

\begin{equation}
= n
\end{equation}

3-Varianza

$$E[x^2]-(E[x])^2$$

\begin{equation}
E[x] = \int_{0}^{\infty} x^2f(x) dx = \int_{0}^{\infty}x^2ax^{\frac{n}{2}-1}  e^{\frac{x}{2}} dx
\end{equation}

\begin{equation}
= a \int_{0}^{\infty}x^{\frac{n}{2}+1}  e^{\frac{x}{2}} dx 
\end{equation}


\begin{equation}
= a \left[-x^{\frac{n}{2}+1} 2 exp(-\frac{n}{2}) \right] + \int_{0}^{\infty} (\frac{n}{2}+1) x^{\frac{n}{2}} exp(-\frac{x}{2}) dx
\end{equation}

\begin{equation}
= a\left[ (n+2) \int_{0}^{\infty}  x^{\frac{n}{2}} exp(-\frac{x}{2}) dx\right] 
\end{equation}


\begin{equation}
=a(n+2) \int_{0}^{\infty} x^{\frac{n}{2}} exp(-\frac{x}{2}) dx
\end{equation}

Volviendo a integrar por partes se obtiene: 

\begin{equation}
= (n+2)n \int_{0}^{\infty} a x^{\frac{n}{2}-1} exp(-\frac{x}{2}) dx
\end{equation}

\begin{equation}
= (n+2)n \int_{0}^{\infty}af(x) dx
\end{equation}

\begin{equation}
= (n+2)n
\end{equation}

Luego:
$$(E[x])^2 = n^2$$

Se tiene pues: 
\begin{equation}
Var = E[x^2] - (E[x])^2 = (n+2)n - n^2 = 2n
\end{equation}




Distribución investigada: DIstribución de Bernoulli

Es la distribucíón que se emplea cuando teniendose una prueba o experimento aleatorio se desea determinar si será exitoso o fallido. Se utiliza "p" para representar la probabilidad de éxito y "q = 1-p" como la probabilidad de fracaso. Es entonces conveniente utilizar una representación de variable discreta para definir el suceso, donde x = 0 si no ocurre el evento y x = 1 en el caso contrario. 

 \begin{cases} 
      p & si x = 1 \\
      1-p & si x=0 \\
      0 & si x \ne R 
 \end{cases}
 
 
 1-CFD:
 
 \begin{equation}
 f(x) = \sum_{x \in R} f(x) = \sum_{x \in R} p^{x}(1-p)^{x} 
 \end{equation}
 
 2-Media:
 
 \begin{equation}
 E[x] = \sum_{x\in R} xf(x)
 \end{equation}
 
 \begin{equation}
 = 1 \cdot p(x)(1) + 0 \cdot p(x)(0)  
 \end{equation}
 
 \begin{equation}
 = 1 \cdot p + 0 \cdot (1-p) 
 \end{equation}
 
 \begin{equation}
 = p
 \end{equation}
 
 3-Varianza
 
 \begin{equation}
 E[x^2] = \sum_{x\in R} x^2f(x) 
 \end{equation}
 
 \begin{equation}
 =  1^2 \cdot p(x)(1) + 0 \cdot p(x)(0)
 \end{equation}
 
 \begin{equation}
 = p
 \end{equation}
 
 Entonces: 
 
 \begin{equation}
 E[x^2]-(E[x])^2 = p - p^2 = p(1-p)
 \end{equation}
